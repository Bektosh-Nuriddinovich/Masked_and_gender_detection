```python
import numpy as np
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.image import img_to_array
import cv2
import cvlib as cv
```

```python
model1 = load_model("C:/Users/Bek/Downloads/Telegram Desktop/model_aug.h5")
model2 = load_model("C:/Users/Bek/Downloads/Telegram Desktop/maska.h5")
```

```python
label1=["masked","no_masked"]
label2 =["Man","Woman"]
video = cv2.VideoCapture("C:/Users/Bek/Downloads/Telegram Desktop/1.mp4")
```

```python
while video.isOpened():
    _,frame = video.read()
    face,confindence = cv.detect_face(frame)
    gender,confindence2 = cv.detect_gender(frame)
    for index,yuz in enumerate(face):
        (startX,startY,endX,endY) = yuz[0],yuz[1],yuz[2],yuz[3]
        
        face_np = np.copy(frame[startY:endY,startX:endX])
        
        if face_np.shape[0] < 10 or face_np.shape[1] < 10:
            continue
        
        face_np = cv2.resize(face_np,(224,224))
        face_np = face_np.astype("float") / 255.0
        face_np = img_to_array(face_np)
        face_np = np.expand_dims(face_np,axis=0)

        predict = model1.predict(face_np)[0]
        predict1 = model2.predict(face_np)[0]
        print("predict=",predict)
        print("predict1=",predict1)

        index = np.argmax(predict)
        index2 = np.argmax(predict1)
        print("index = ",index)
        print("index2 = ",index2)
        label = label1[index]
        label3 = label2[index2]
        print("label = ",label)
        print("label2 = ",label3)

        if label == 'masked':
            color = (0,255,0)
        else:
            color = (0,0,255)
            
        label = f"{label} {np.around(predict[index]*100,2)} {label3}"
        
        if startY-10 > 10:
            Y=startY-10
        else:
            Y=startY+10
        
        cv2.rectangle(frame,(startX,startY),(endX,endY),color,2)
        cv2.putText(frame,label,(startX,Y),cv2.FONT_HERSHEY_SIMPLEX,0.7,color,2)
      
    cv2.imshow("Video",frame)
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
video.release()
cv2.destroyAllWindows()
```
